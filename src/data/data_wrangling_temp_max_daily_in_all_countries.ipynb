```python
print("Start")

import pandas as pd

df = pd.read_csv("../../../dataset-sis-energy-derived-reanalysis-temperature-country_daily/temp_country_daily.csv", sep=";")
```

```python
# data.head()
# data.info()
```

```python
# convert column holding the date from Object to Datetime
df["Date"] = pd.to_datetime(df["Date"])

# resample from daily to week data and aggregate on mean temperature values over the week
# not resample("7D") as before as overflow to following year happens
# df = df.resample("1W", on="Date").max()

# convert temperature values from Kelvon to Celsius for all columns except the date column
df.loc[:, df.columns != 'Date'] = df.loc[:, df.columns != 'Date'] - 273.15
```

```python
# reset index from date value to integer 
#df = df.reset_index()
# get weeknumber as separate column
df["weeknumber"] = df.Date.dt.isocalendar().week
# get year number as separate column
df["year"] = df.Date.dt.isocalendar().year
```

```python
# view data
df.head()
```

```python
# group df by year and weeknumber and use max daily temperature
df_grouped = df.groupby(by=["year", "weeknumber"], group_keys=False).max()
df_grouped.head()
```

```python
# get weeknumber as separate column
df_grouped["weeknumber"] = df_grouped.Date.dt.isocalendar().week
# get year number as separate column
df_grouped["year"] = df_grouped.Date.dt.isocalendar().year
```

```python
# save result to a csv file
df_grouped.to_json('temp_country_weekly_max_daily_in_week_all_countries.json', orient='records')
```
